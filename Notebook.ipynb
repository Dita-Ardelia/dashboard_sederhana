```python

# Fungsi untuk menghitung rata-rata harga produk per kategori
def calculate_average_price(merged_df):
    if "product_category_name" not in merged_df.columns or "price" not in merged_df.columns:
        raise ValueError("Kolom 'product_category_name' atau 'price' tidak ditemukan dalam dataset!")

    category_price = merged_df.groupby("product_category_name")["price"].mean().reset_index()
    category_price.columns = ["category", "average_price"]
    category_price = category_price.sort_values(by="average_price", ascending=False).head(10)  # Top 10 kategori
    return category_price

# Fungsi untuk visualisasi data
def plot_category_price(category_price):
    plt.figure(figsize=(12, 6))

    sns.barplot(
        y=category_price["category"],
        x=category_price["average_price"],
        hue=category_price["category"],  # Memperbaiki warning Seaborn
        palette="Oranges_r",
        legend=False  # Mencegah legend berlebihan
    )

    plt.title("Rata-rata Harga Produk per Kategori (Top 10)")
    plt.xlabel("Rata-rata Harga (USD)")
    plt.ylabel("Kategori Produk")
    plt.show()

# Memuat dataset jika belum ada
try:
    products_df
    order_df
except NameError:
    print("Dataset belum dimuat, memuat data sekarang...")
    products_df = pd.read_csv("products.csv")
    order_df = pd.read_csv("order.csv")
    print("File berhasil dimuat!")

# Menggabungkan dataset dan melakukan analisis
merged_df = merge_data(products_df, order_df)
category_price = calculate_average_price(merged_df)
plot_category_price(category_price)
```

```python
try:
    order_df
    products_df
except NameError:
    print("Dataset belum dimuat, memuat data sekarang...")
    order_df = pd.read_csv("order.csv")  # Sesuaikan dengan nama file yang benar
    products_df = pd.read_csv("products.csv")
    print("File berhasil dimuat!")

# Merge datasets pada 'product_id'
merged_df = order_df.merge(products_df, on="product_id", how="left")

# Pastikan kolom yang dibutuhkan ada
if "product_category_name" not in merged_df.columns or "price" not in merged_df.columns:
    raise ValueError("Kolom 'product_category_name' atau 'price' tidak ditemukan dalam dataset!")

# Menghitung total pendapatan per kategori
category_revenue = merged_df.groupby("product_category_name")["price"].sum().reset_index()

# Ubah nama kolom
category_revenue.columns = ["category", "total_revenue"]

# Hapus kategori dengan nilai NaN
category_revenue = category_revenue.dropna()

# Urutkan berdasarkan pendapatan tertinggi
category_revenue = category_revenue.sort_values(by="total_revenue", ascending=False).head(10)  # Top 10 kategori

# Visualisasi menggunakan bar chart
plt.figure(figsize=(12, 6))

sns.barplot(
    y=category_revenue["category"],
    x=category_revenue["total_revenue"],
    hue=category_revenue["category"],  # Memperbaiki warning Seaborn
    palette="Greens_r",
    legend=False  # Menghindari legenda yang tidak diperlukan
)

# Tambahkan judul dan label
plt.title("Pendapatan Tertinggi per Kategori Produk (Top 10)")
plt.xlabel("Total Pendapatan (USD)")
plt.ylabel("Kategori Produk")

# Tampilkan plot
plt.show()
```

**Insight:**
- Menampilkan Pendapatan Tertinggi per Kategori Produk
